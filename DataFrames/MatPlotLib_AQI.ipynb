```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import glob
import scipy.stats as st
import datetime as dt
from datetime import datetime
import matplotlib.dates as mdates
import numpy

# Study data files
pm25_data_path = "../Resources/aqi.csv"
ozone_data_path = "../Resources/ozone.csv"
covid_path = "../Resources/mn-c19.csv"
#Traffic_Data = "data/MNDOT.csv"

# Read the PM2.5 data and the Ozone Data
pm25_data = pd.read_csv(pm25_data_path)
ozone_data = pd.read_csv(ozone_data_path)

#Combine all AQI files
Combined_AQI_csv=pd.merge(pm25_data,ozone_data,on="Date",how="left")
#Export to CSV
Combined_AQI_csv.to_csv( "../Resources/Combined_AQI_csv.csv", index=False)
#Combined_AQI_csv.head()

# add covid data
combine_path = "../Resources/Combined_AQI_csv.csv"
combine_data = pd.read_csv(combine_path)
covid_data = pd.read_csv(covid_path)
combined_aqi_covid = pd.merge(combine_data, covid_data,on="Date",how="left")

#export to csv
combined_aqi_covid.to_csv("../Resources/combined_aqi_covid.csv", index=False)

#combined_aqi_covid.head()
```

```python
#Create plot tracking AQI change over time
PM25 = Combined_AQI_csv["Daily Mean PM2.5 Concentration"]
dates = Combined_AQI_csv['Date']

plt.figure(figsize=(15,3))

ax = plt.gca()
dates = [datetime.strptime(d, "%m/%d/%Y") for d in dates]

ax.set_xlim([dates[0], dates[-1]])
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3,6,9), bymonthday=-1))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

plt.plot(dates,PM25,color="g")
plt.title("PM2.5 Concentration Over Time")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.ylabel("PM2.5 Concentration (ug/m3 LC)")

#PM2.5 Concentration Limits
plt.ylim(0,75)
#Date Limits
plt.xlim(pd.Timestamp('2015-01-01'), pd.Timestamp('2020-11-01'))

#Print
plt.savefig('../Images/AQI-AllYears.jpg')
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# Based on PM2.5 over Time

PM25 = Combined_AQI_csv["Daily Mean PM2.5 Concentration"]
dates = Combined_AQI_csv['Date']

plt.figure(figsize=(15,3))
# Regression
m_slope, m_int, m_r, m_p, m_std_err = st.linregress(Combined_AQI_csv.index, PM25)
regress_values = Combined_AQI_csv.index * m_slope + m_int
plt.plot(Combined_AQI_csv.index,regress_values,"r-",color="lightskyblue")
line_eq = "y = " + str(round(m_slope,2)) + "x + " + str(round(m_int,2))
# scatterplot (comment out 'dates =' and 'plt.xlim' below for regress to work, but then the dates won't work)
ax = plt.gca()
#dates = [datetime.strptime(d, "%m/%d/%Y") for d in dates]
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_minor_locator(mdates.MonthLocator(bymonth=(3,6,9), bymonthday=-1))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))

plt.scatter(dates, PM25, marker="^", facecolors="xkcd:gray", edgecolors="black")
plt.annotate(line_eq,(50,40),fontsize=15,color="steelblue")
plt.title("PM 2.5 Regression Analysis 2015 - Present")
plt.xlabel("Time")
#plt.xlim([dates[0], dates[-1]])
plt.xticks(rotation=45)
plt.ylabel("PM2.5 Concentration (ug/m3 LC)")
print(f"The r-squared is: {m_r**2}")
plt.savefig('../Images/AQI-regress.jpg')
plt.show()


#Correlation Coefficient
#W=Combined_AQI_csv["Date"]
#A=Combined_AQI_csv["Daily Mean PM2.5 Concentration"]
```

```python
#combined covid aqi path and reading in csv
combined_covid_path = "../Resources/combined_aqi_covid_dates.csv"
covid_df = pd.read_csv(combined_covid_path)

covid_df.head()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y')
#x_ticks = ["Mar 20"]
# make a plot
ax.plot(covid_df.Date, covid_df.DAILY_AQI_VALUE_x, label="AQI Particulates")
#append above if needed: , index=months, color="red", marker="o")
#dates = aqi_covid_5yrdf['Date']


#ax.set_xticks(mdates.MonthLocator(bymonth=(3,6,9), bymonthday=-1))


ax.legend()
# set x-axis label
ax.set_xlabel("2015-2020",fontsize=14)
# set y-axis label
ax.set_ylabel("AQI pm2.5",color="red",fontsize=14)
```

```python
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(covid_df.Date, covid_df.positive,color="blue",marker="o",label="Covid Cases")
ax2.set_ylabel("MN Covid Cases",color="blue",fontsize=14)
plt.show()

ax2.xaxis.set_major_locator(years)
ax2.xaxis.set_major_formatter(years_fmt)
ax2.xaxis.set_minor_locator(months)
ax2.legend()
# save the plot as a file
fig.savefig('../Images/aqi_5yr.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
